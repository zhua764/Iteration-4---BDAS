```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rcParams
```

```python
pdf = pd.read_csv("./balanced.csv")
pdf_rf = pd.read_csv("./rf__predict.csv")
pdf_lr = pd.read_csv("./lr__predict.csv")
```

###### 8.2 Visualize the data, result, models and patterns

```python
rcParams['figure.figsize'] = 11, 8
sns.countplot(x="age", hue="cardio", data=pdf, palette="Set2");
```

```python
df_melt = pd.melt(frame=pdf, value_vars=['weight'], id_vars=['cardio'])
plt.figure(figsize=(12, 10))
ax = sns.violinplot(
     x='variable', 
     y='value', 
     hue='cardio', 
     split=True,
     data=df_melt, 
     scale='count',
     scale_hue=False,
     palette="Set2");
```

```python
df_categorical = pdf.iloc[:, 2:5]
sns.countplot(x="variable", hue="value",data= pd.melt(df_categorical));
```

```python
sns.countplot(x='cardio',data=pdf_rf,linewidth=2,
               edgecolor=sns.color_palette("dark", 1))
plt.title('Random Forest')
```

```python
sns.countplot(x='cardio',data=pdf_lr,linewidth=2,
               edgecolor=sns.color_palette("dark", 1))
plt.title('Logistic Regression')
```
